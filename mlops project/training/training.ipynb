## Preprocessing pipeline

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
```

```python
train_data = train_df.copy()
test_data = test_df.copy()

```

```python
y_train = train_data['Close'].values
X_train = train_data.drop('Close', axis = 1).values
```

```python
y_test = test_data['Close'].values.reshape(-1,1)
X_test = test_data.drop('Close', axis = 1).values
```

```python
# Create the pipeline with scaler and SVR
rope = Pipeline([('scaler', StandardScaler()), ('svr', SVR())])

# Define the hyperparameter grid to search
param_grid = [
    {'svr__kernel': ['linear'], 'svr__C': [1, 10, 100, 1000]},
    {'svr__kernel': ['rbf'], 'svr__C': [1, 10, 100, 1000], 'svr__gamma': [0.001, 0.0001], 'svr__epsilon':[0.01,1]},
]
# Create the GridSearchCV object
grid_search = GridSearchCV(rope, param_grid=param_grid, cv=5)

# Fit the GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Print the best hyperparameters
print("Best parameters:", grid_search.best_params_)

# Get the best model from the search
best_model = grid_search.best_estimator_

# Make predictions using the best model
y_p = best_model.predict(X_test)
```

```python
pipe = Pipeline([('scaler', StandardScaler()), ('svr', SVR(C= 1000, kernel =  'linear'))])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test).reshape(-1,1)
```

```python
MAPE = np.average(np.abs((y_test - y_pred) / y_test)) * 100
print(f"MAPE: {MAPE:.2f}%")
print()
```

```python

print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print()
```

```python


data = pd.DataFrame({'Date': test_data.index.values, 'Bitcoin Price': y_test.flatten(), 'Predicted Bitcoin Price': y_pred.flatten()})

# Create a line plot using Plotly Express
fig = px.line(data, x='Date', y=['Bitcoin Price', 'Predicted Bitcoin Price'],
              labels={'value': 'Price'},
              line_shape='linear',
              title='SVR')

# Update the y-axis tick format to include commas for thousands separator
fig.update_layout(yaxis_tickformat=',d')

# Show the figure
fig.show()
```
